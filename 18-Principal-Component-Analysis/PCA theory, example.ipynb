1. Import Libraries and Load the Data

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Example dataset
data = {
    'Feature1': [2.5, 0.5, 2.2, 1.9, 3.1],
    'Feature2': [2.4, 0.7, 2.9, 2.2, 3.0]
}
df = pd.DataFrame(data)
```

2. Standardizing the Data

```python
# Standardizing the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

print("Scaled Data:")
print(scaled_data)
```

3. Applying PCA

```python
# Initialize PCA to compute all components
pca = PCA(n_components=2)  # We can change n_components to reduce dimensions
pca_result = pca.fit_transform(scaled_data)

# Display principal components
print("PCA Components:")
print(pca.components_)

print("PCA Transformed Data:")
print(pca_result)

# Explained variance
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
```

4. Visualizing the Explained Variance

```python
import matplotlib.pyplot as plt

# Plot explained variance ratio
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Components')
plt.show()
```
