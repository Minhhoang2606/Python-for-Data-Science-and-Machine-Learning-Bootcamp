## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

## The Data

Let's work with the cancer data set again since it had so many features.

```python
# Load the dataset
cancer = load_breast_cancer()

# Check the keys available in the dataset
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
# Create a DataFrame
df = pd.DataFrame(data=cancer['data'], columns=cancer['feature_names'])

# Display the first few rows
df.head()
```

## PCA Visualization

As we've noticed before it is difficult to visualize high dimensional data, we can use PCA to find the first two principal components, and visualize the data in this new, two-dimensional space, with a single scatter-plot. Before we do this though, we'll need to scale our data so that each feature has a single unit variance.

```python
# Initialize the scaler
scaler = StandardScaler()

# Scale the data
scaled_data = scaler.fit_transform(df)
```

```python
# Display dataset information
print(f"Number of samples: {scaled_data.shape[0]}")
print(f"Number of features: {scaled_data.shape[1]}")
```

PCA with Scikit Learn uses a very similar process to other preprocessing functions that come with SciKit Learn. We instantiate a PCA object, find the principal components using the fit method, then apply the rotation and dimensionality reduction by calling transform().

We can also specify how many components we want to keep when creating the PCA object.

```python
from sklearn.decomposition import PCA

# Initialize PCA and specify the number of components
pca = PCA(n_components=2)
```

```python
# Fit PCA to the scaled data and transform it
pca_data = pca.fit_transform(scaled_data)

# Check the shape of the transformed data
print(f"Shape of transformed data: {pca_data.shape}")
```

```python
# Display the explained variance ratio
print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
```

```python
# Import matplotlib for visualization
import matplotlib.pyplot as plt

# Create a scatter plot of the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=cancer['target'], cmap='plasma', edgecolor='k', s=50)

# Label the axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('2D Visualization of PCA Transformed Data')
plt.colorbar(label='Target (0: Malignant, 1: Benign)')
plt.show()
```

```python
from matplotlib.gridspec import GridSpec

# Create a figure with GridSpec for better control over layout
fig = plt.figure(figsize=(16, 6))
gs = GridSpec(1, 3, width_ratios=[1, 1, 0.05], wspace=0.3)

# Original dataset subplot
ax0 = fig.add_subplot(gs[0, 0])
scatter_original = ax0.scatter(
    df['mean radius'],
    df['mean texture'],
    c=cancer['target'],
    cmap='plasma',
    edgecolor='k',
    s=50
)
ax0.set_title('Original Dataset (Selected Features)')
ax0.set_xlabel('Mean Radius')
ax0.set_ylabel('Mean Texture')

# PCA-transformed dataset subplot
ax1 = fig.add_subplot(gs[0, 1])
scatter_pca = ax1.scatter(
    pca_data[:, 0],
    pca_data[:, 1],
    c=cancer['target'],
    cmap='plasma',
    edgecolor='k',
    s=50
)
ax1.set_title('PCA-Transformed Dataset')
ax1.set_xlabel('First Principal Component')
ax1.set_ylabel('Second Principal Component')

# Colorbar in the third grid space
cbar_ax = fig.add_subplot(gs[0, 2])
cbar = fig.colorbar(scatter_pca, cax=cbar_ax)
cbar.set_label('Target (0: Malignant, 1: Benign)')

plt.show()
```

```python
pca.components_
```

```python
# Create a DataFrame for the components
component_df = pd.DataFrame(pca.components_, columns=cancer['feature_names'])

# Plot a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(component_df, cmap='plasma', annot=False)
plt.title('Feature Contributions to Principal Components')
plt.xlabel('Features')
plt.ylabel('Principal Components')
plt.show()
```

```python
pca.fit(scaled_data)
```

Now we can transform this data to its first 2 principal components.

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

Great! We've reduced 30 dimensions to just 2! Let's plot these two dimensions out!

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

Clearly by using these two components we can easily separate these two classes.

## Interpreting the components 

Unfortunately, with this great power of dimensionality reduction, comes the cost of being able to easily understand what these components represent.

The components correspond to combinations of the original features, the components themselves are stored as an attribute of the fitted PCA object:

```python
pca.components_
```

In this numpy matrix array, each row represents a principal component, and each column relates back to the original features. we can visualize this relationship with a heatmap:

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```

This heatmap and the color bar basically represent the correlation between the various feature and the principal component itself.
